# Convert JSON Data

```python
# Load json data and turn into dictionaries

with open(data_path, 'r', encoding='utf-8') as f:
    data = json.loads(f.read())
```

```python
# Normalized all
normal_df = pd.json_normalize(data)
# Enlarge column width
normal_df.head().style.set_properties(subset=['source'], **{'width': '100px'})
```

### Category Columns

```python
normal_df['category']
```

```python
empty_category = normal_df['category'].apply(lambda x: True if len(x) == 0 else False)
empty_category[empty_category]
```

```python
# Number of no category
empty_category[empty_category].sum()
```

```python
# Fill the empty category
def fill_empty_cat(x):
    if len(x) == 0:
        x.append('Uncategorized')
    return x

normal_df['category'] = normal_df['category'].apply(lambda x: fill_empty_cat(x))
```

```python
normal_df['category']
```

```python
category_list = []
for c in normal_df['category']:
    for x in c:
        category_list.append(x)
cat_series = pd.Series(category_list)
```

```python
pd.set_option("display.max_rows", 38)
cat_series.value_counts().to_frame()
```

```python
# Download the document files using wget and the non_blank_doc dataframe
download = True # Add download toggler to prevent accidental document download
DATASET_DIR = "../../adil-dataset/"

if download:
    import requests
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36"}
    series_df = pd.concat([non_blank_doc_df['id'], non_blank_doc_df['document']], axis=1)
    for idx, docs in series_df.iterrows():
        doc_id = docs['id']
        res = requests.get(docs['document'][0], stream=True, headers=headers)
        try:
            with open(DATASET_DIR+doc_id+".0.pdf", "wb") as pdf:
                pdf.write(res.content)
            print("Downloaded File {}.pdf".format(doc_id))
        except Exception as e:
            print(e)
            print("Unable to download file {} with status code {}".format(doc_id, res.status_code))
```
